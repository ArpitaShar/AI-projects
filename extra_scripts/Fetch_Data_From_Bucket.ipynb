```python
import oci
import os
```

```python

config = oci.config.from_file("~/.oci/config", "DEFAULT")

object_storage = oci.object_storage.ObjectStorageClient(config)
```

```python

# Set your variables here
namespace = 'bmmeqns3hjh7'           # e.g., 'axax7q9ehabc'
bucket_name = 'ReportGenTool_bucket_MGNREGA'       # e.g., 'my-data-bucket'
object_name = 'MG_WRKDTLCMP_2223_to_2526.csv'          # e.g., 'dataset.csv'
download_path = os.path.join('/home/ubuntu/workspace/Notebooks/Mohit/Data', object_name)  # e.g., '/home/user/data/dataset.csv'
```

```python
try:
    response = object_storage.get_object(namespace, bucket_name, object_name)

    # Write to local file
    with open(download_path, 'wb') as f:
        for chunk in response.data.raw.stream(1024 * 1024, decode_content=False):
            f.write(chunk)

    print(f"File '{object_name}' downloaded successfully to '{download_path}'.")

except oci.exceptions.ServiceError as e:
    print(f"Error fetching object: {e}")
```
